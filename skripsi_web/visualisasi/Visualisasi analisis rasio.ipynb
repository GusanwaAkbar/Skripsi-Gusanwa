```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
presisi = pd.read_csv("Tabel Presisi - Sheet1.csv")
recall = pd.read_csv("Tabel Recall - Sheet1.csv")
f1 = pd.read_csv("Tabel F1 Score - Sheet1.csv")
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
cmap = sns.diverging_palette(250, 10, s=90, l=50, as_cmap=True)


sns.heatmap(presisi, annot=True, fmt=".1f", ax = ax, cmap=cmap)
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
cmap = sns.diverging_palette(250, 10, s=90, l=50, as_cmap=True)


sns.heatmap(recall, annot=True, fmt=".1f", ax = ax, cmap=cmap)
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
cmap = sns.diverging_palette(250, 10, s=90, l=50, as_cmap=True)


sns.heatmap(f1, annot=True, fmt=".1f", ax = ax, cmap=cmap)
```
